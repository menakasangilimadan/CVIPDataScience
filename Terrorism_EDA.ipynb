```python
###############EDA for World Terrorism ################
#import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#To ignore some warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the data
data=pd.read_csv("globalterrorismdb_0718dist.csv",encoding='ISO-8859-1')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_info_columns', len(data.columns))
```

```python
#To view the nos of rows and columns
data.shape
```

```python
#As there are many columns .So we can take only necessary columns
terrordata = data[['eventid','iyear','imonth','iday','country_txt','region_txt','provstate','city','latitude','longitude','attacktype1_txt','gname','targtype1_txt','target1','weaptype1_txt','weapdetail','nwound','nkill','motive']]
terrordata.rename(columns = {'eventid':'EventId','iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'Latitude','longitude':'Longitude','attacktype1_txt':'Attacktype','gname':'GroupName','targtype1_txt':'TargetType','target1':'Target','weaptype1_txt':'WeaponType','weapdetail':'WeaponDetail','nwound':'Wound','nkill':'Killed','motive':'Motive'}, inplace = True)
terrordata
```

```python
#droping the some columns which has null values
terrordata_new=terrordata.drop(['State', 'City','Target','WeaponDetail','Motive'], axis=1)
terrordata_new
```

```python
#Before EDA let us divide the categorical values and non categorical values

categorical=terrordata_new.select_dtypes(include=['object']).columns
noncategorical = terrordata_new.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(categorical)
print("Numerical Variables:")
print(noncategorical)
```

```python
countrycount=terrordata_new['Country'].value_counts()
countrycount[:20].plot.bar()
```

```python
city=terrordata_new['Region'].value_counts()
city[:20].plot.bar()
```

```python

Killedcount=terrordata_new[['Country','Attacktype','Killed']] [terrordata_new['Killed'] > 0].groupby(['Country','Attacktype'],as_index=False)
Killedcount=Killedcount.sum().sort_values(by='Killed', ascending=False)[:10]
Killedcount
Killedcount.plot.bar(x="Country",stacked=True,title="Top 10 Coutries with High Killed Rates")
```
